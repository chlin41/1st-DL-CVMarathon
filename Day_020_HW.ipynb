```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
matrix1 = np.random.random(100)
matrix = np.random.random(100)*2-1
matrix2 = np.random.normal(size = 100)
matrix3 = np.random.randn(10,10)
matrix4 = np.random.uniform(low=-1, high=1, size=(100,))
```

```python
plt.figure(figsize=(8,8))
sns.distplot(matrix1)
sns.distplot(matrix)
sns.distplot(matrix2)
sns.distplot(matrix3)
sns.distplot(matrix4)
plt.legend(labels=['matrix1', 'matrix', 'matrix2', 'matrix3', 'matrix4'])
```

```python
matrix = matrix.reshape(10,10)
sns.heatmap(matrix, vmin = (-1), vmax= 1, annot=True, cmap = plt.cm.RdYlBu_r)
```

```python
"""
請用 numpy 建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣並繪製 Heatmap
Hint : numpy.random.random - 均勻分布, 隨機小數
"""
matrix = (np.random.rand(100)*2-1).reshape(10, 10)

plt.figure(figsize=(10,10))
sns.heatmap(matrix, vmin = (-1), vmax= 1, annot=True, cmap = plt.cm.RdYlBu_r)
plt.show()
```

```python
nrow = 1000
ncol = 3
"""
請用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，
並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)
"""
matrix = (np.random.rand(3000)*2-1).reshape(1000, 3)

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)


plt.show()
```

```python
nrow = 1000
ncol = 3
"""
請用 numpy 建立一個 1000 x 3, 數值分布為常態分佈的矩陣，
並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)
"""
matrix = (np.random.randn(3000)).reshape(1000,3)

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False,
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
